```python
import seaborn as sns
```

```python
file_path = "../tools/in/0001.txt"
```

```python
with open(file_path) as f:
    n, m, k = map(int, f.readline().split())
    x = []
    y = []
    for i in range(n):
        xi, yi = map(int, f.readline().split())
        x.append(xi)
        y.append(yi)
    
    u = []
    v = []
    w = []
    for i in range(m):
        ui, vi, wi = map(int, f.readline().split())
        u.append(ui)
        v.append(vi)
        w.append(wi)
        
    a = []
    b = []
    for i in range(k):
        ai, bi = map(int, f.readline().split())
        a.append(ai)
        b.append(bi)
```

```python
sns.scatterplot(x=a, y=b)
# sns.scatterplot(x=x, y=y)
```
